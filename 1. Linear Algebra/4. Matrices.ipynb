# Matrices in Linear Algebra for AI/ML

## Introduction and Relevance
**Definition**: A matrix is a rectangular array of numbers arranged in rows and columns, denoted as $\mathbf{A} \in \mathbb{R}^{m \times n}$, where $m$ is the number of rows and $n$ is the number of columns. In AI/ML, matrices represent datasets, transformations, or model parameters.

**Mathematical Representation**: A matrix $\mathbf{A}$ is written as:
$$
\mathbf{A} = \begin{bmatrix}
a_{11} & a_{12} & \dots & a_{1n} \\
a_{21} & a_{22} & \dots & a_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{m1} & a_{m2} & \dots & a_{mn}
\end{bmatrix}
$$
where $a_{ij}$ is the element in the $i$-th row and $j$-th column.

**Examples**:
- 2x2 matrix: $\mathbf{A} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$
- 3x2 matrix: $\mathbf{B} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \\ 5 & 6 \end{bmatrix}$
- ML context: A dataset with $m$ samples and $n$ features is an $m \times n$ matrix, where each row is a feature vector (e.g., pixel values in MNIST).

**Relevance in Mathematics**:
- **Linear Transformations**: Matrices represent operations like rotations, scaling, or projections.
- **Systems of Equations**: Solve multiple linear equations simultaneously.
- **Generalization**: Extend vector operations to higher dimensions, enabling concepts like eigenvalues and SVD.
- **Structural Analysis**: Study properties like rank, determinant, or invertibility.

**Relevance in AI/ML**:
- **Data Storage**: Represent datasets (rows as samples, columns as features).
- **Model Parameters**: Store weights in neural networks or coefficients in regression.
- **Transformations**: Apply linear transformations in neural network layers or preprocessing.
- **Optimization**: Facilitate gradient-based updates and loss computations.

**Applications Overview**:
- **Data Preprocessing**: Normalizing or combining feature matrices.
- **Neural Networks**: Transforming inputs via weight matrices.
- **NLP**: Representing embeddings or attention scores in transformers.
- **Computer Vision**: Representing images or applying transformations.
- **Optimization**: Updating weight matrices in gradient descent.
- **Recommendation Systems**: Factorizing user-item matrices.

### 2.1 Matrix Representation
**Formula**: An $m \times n$ matrix is a grid of numbers with $m$ rows and $n$ columns, represented as $\mathbf{A} = [a_{ij}]$, where $a_{ij}$ is the element at row $i$ and column $j$.

**Explanation**: Matrices generalize vectors to two dimensions, enabling structured representation of data (e.g., datasets, images) or transformations (e.g., neural network layers).

**Example**: A 2x2 matrix $\mathbf{A} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$ or a 3x2 dataset matrix representing three samples with two features.

**Why Necessary**: Provides a compact way to represent multidimensional data or transformations, enabling efficient ML computations.

```python
# Matrix Representation
import numpy as np

# Define a 2x2 matrix
A = np.array([[1, 2], [3, 4]])
print("2x2 Matrix:\n", A)

# Define a 3x2 dataset matrix (3 samples, 2 features)
dataset = np.array([[1, 2], [3, 4], [5, 6]])
print("3x2 Dataset Matrix:\n", dataset)
```

### 2.2 Matrix Addition
**Formula**: For matrices $\mathbf{A}$ and $\mathbf{B}$ of size $m \times n$, their sum is $\mathbf{C} = \mathbf{A} + \mathbf{B}$, where $c_{ij} = a_{ij} + b_{ij}$.

**Explanation**: Combines matrices element-wise, producing a new matrix of the same size. Analogous to vector addition but in 2D.

**Example**: Given $\mathbf{A} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$ and $\mathbf{B} = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}$, compute $\mathbf{A} + \mathbf{B}$:
- Step 1: Add corresponding elements:
  $$
  \mathbf{A} + \mathbf{B} = \begin{bmatrix} 1+5 & 2+6 \\ 3+7 & 4+8 \end{bmatrix} = \begin{bmatrix} 6 & 8 \\ 10 & 12 \end{bmatrix}
  $$

**Why Necessary**: Aggregates information, used in ML for summing gradients or combining feature matrices.

**ML Use-Case**: In batch gradient descent, gradient matrices from multiple samples are summed to update weights.

```python
# Matrix Addition Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define matrices as collections of vectors
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = A + B

# Visualize as vector transformations
vectors = np.array([[1, 0], [0, 1]])  # Standard basis
A_vectors = A @ vectors.T # Matrix-vector multiplication , the @ operator is used for matrix multiplication in numpy
B_vectors = B @ vectors.T
C_vectors = C @ vectors.T

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, A_vectors[0, 0], A_vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='blue', label='A Basis 1')
plt.quiver(0, 0, A_vectors[0, 1], A_vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5, label='A Basis 2')
plt.quiver(0, 0, C_vectors[0, 0], C_vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='green', label='A+B Basis 1')
plt.quiver(0, 0, C_vectors[0, 1], C_vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='green', alpha=0.5, label='A+B Basis 2')

plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Matrix Addition as Vector Transformation')
plt.legend()
plt.show()

print("Matrix Addition Result:\n", C)
```

### 2.3 Scalar Multiplication
**Formula**: For a matrix $\mathbf{A}$ and scalar $c$, the result is $\mathbf{B} = c \mathbf{A}$, where $b_{ij} = c a_{ij}$.

**Explanation**: Scales each element of the matrix, stretching or shrinking its values.

**Example**: Given $\mathbf{A} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$ and $c = 2$, compute $c \mathbf{A}$:
- Step 1: Multiply each element:
  $$
  c \mathbf{A} = \begin{bmatrix} 2 \cdot 1 & 2 \cdot 2 \\ 2 \cdot 3 & 2 \cdot 4 \end{bmatrix} = \begin{bmatrix} 2 & 4 \\ 6 & 8 \end{bmatrix}
  $$

**Why Necessary**: Adjusts matrix magnitude, used in ML for scaling gradients or normalizing features.

**ML Use-Case**: In gradient descent, the learning rate scales the gradient matrix: $\mathbf{W} \leftarrow \mathbf{W} - \eta \nabla L$.

```python
# Scalar Multiplication Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define matrix and scalar
A = np.array([[1, 2], [3, 4]])
c = 2
B = c * A

# Visualize as vector transformations
vectors = np.array([[1, 0], [0, 1]])  # Standard basis
A_vectors = A @ vectors.T
B_vectors = B @ vectors.T

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, A_vectors[0, 0], A_vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='blue', label='A Basis 1')
plt.quiver(0, 0, A_vectors[0, 1], A_vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5, label='A Basis 2')
plt.quiver(0, 0, B_vectors[0, 0], B_vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='purple', label='2A Basis 1')
plt.quiver(0, 0, B_vectors[0, 1], B_vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='purple', alpha=0.5, label='2A Basis 2')

plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Scalar Multiplication as Vector Transformation')
plt.legend()
plt.show()

print("Scalar Multiplication Result:\n", B)
```

### 2.4 Matrix Multiplication
**Formula**: For matrices $\mathbf{A} \in \mathbb{R}^{m \times n}$ and $\mathbf{B} \in \mathbb{R}^{n \times p}$, their product is $\mathbf{C} = \mathbf{A} \mathbf{B} \in \mathbb{R}^{m \times p}$, where $c_{ij} = \sum_{k=1}^n a_{ik} b_{kj}$.

**Explanation**: Combines rows of $\mathbf{A}$ with columns of $\mathbf{B}$, representing the composition of linear transformations.

**Example**: Given $\mathbf{A} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$ and $\mathbf{B} = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}$, compute $\mathbf{A} \mathbf{B}$:
- Step 1: Compute each element of $\mathbf{C}$:
  $$
  c_{11} = 1 \cdot 5 + 2 \cdot 7 = 19, \quad c_{12} = 1 \cdot 6 + 2 \cdot 8 = 22
  $$
  $$
  c_{21} = 3 \cdot 5 + 4 \cdot 7 = 43, \quad c_{22} = 3 \cdot 6 + 4 \cdot 8 = 50
  $$
  Result: $\mathbf{C} = \begin{bmatrix} 19 & 22 \\ 43 & 50 \end{bmatrix}$

**Why Necessary**: Applies transformations, critical for neural network layers and data projections.

**ML Use-Case**: In neural networks, matrix multiplication computes $\mathbf{W} \mathbf{X}$ for a layers output.

```python
# Matrix Multiplication Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = np.dot(A, B)

# Visualize as transformation of basis vectors
vectors = np.array([[1, 0], [0, 1]])  # Standard basis
A_vectors = A @ vectors.T
C_vectors = C @ vectors.T

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, vectors[0, 0], vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='black', label='Original Basis 1')
plt.quiver(0, 0, vectors[0, 1], vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='black', alpha=0.5, label='Original Basis 2')
plt.quiver(0, 0, A_vectors[0, 0], A_vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='blue', label='A Basis 1')
plt.quiver(0, 0, A_vectors[0, 1], A_vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5, label='A Basis 2')
plt.quiver(0, 0, C_vectors[0, 0], C_vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='green', label='AB Basis 1')
plt.quiver(0, 0, C_vectors[0, 1], C_vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='green', alpha=0.5, label='AB Basis 2')

plt.xlim(-1, 25)
plt.ylim(-1, 25)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Matrix Multiplication as Transformation')
plt.legend()
plt.show()

print("Matrix Multiplication Result:\n", C)
```

### 2.5 Matrix Transpose
**Formula**: For a matrix $\mathbf{A} \in \mathbb{R}^{m \times n}$, its transpose $\mathbf{A}^T \in \mathbb{R}^{n \times m}$ has elements $(\mathbf{A}^T)_{ij} = a_{ji}$.

**Explanation**: Flips the matrix over its main diagonal, swapping rows and columns.

**Example**: Given $\mathbf{A} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$, compute $\mathbf{A}^T$:
- Step 1: Swap rows and columns:
  $$
  \mathbf{A}^T = \begin{bmatrix} 1 & 3 \\ 2 & 4 \end{bmatrix}
  $$

**Why Necessary**: Aligns matrices for operations like dot products or gradient computations.

**ML Use-Case**: In backpropagation, $\mathbf{W}^T$ is used to compute gradients.

```python
# Matrix Transpose Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define matrix
A = np.array([[1, 2], [3, 4]])
A_T = A.T

# Visualize as transformation
vectors = np.array([[1, 0], [0, 1]])  # Standard basis
A_vectors = A @ vectors.T
A_T_vectors = A_T @ vectors.T

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, A_vectors[0, 0], A_vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='blue', label='A Basis 1')
plt.quiver(0, 0, A_vectors[0, 1], A_vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5, label='A Basis 2')
plt.quiver(0, 0, A_T_vectors[0, 0], A_T_vectors[1, 0], angles='xy', scale_units='xy', scale=1, color='red', label='A^T Basis 1')
plt.quiver(0, 0, A_T_vectors[0, 1], A_T_vectors[1, 1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, label='A^T Basis 2')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Matrix Transpose as Transformation')
plt.legend()
plt.show()

print("Matrix Transpose Result:\n", A_T)
```

### 2.6 Matrix Inverse
**Formula**: For a square matrix $\mathbf{A} \in \mathbb{R}^{n \times n}$, its inverse $\mathbf{A}^{-1}$ satisfies $\mathbf{A} \mathbf{A}^{-1} = \mathbf{I}$, where $\mathbf{I}$ is the identity matrix. Exists if $\det(\mathbf{A}) \neq 0$.

**Explanation**: Reverses a linear transformation, solving systems like $\mathbf{A} \mathbf{x} = \mathbf{b}$ as $\mathbf{x} = \mathbf{A}^{-1} \mathbf{b}$.

**Example**: Given $\mathbf{A} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$, compute $\mathbf{A}^{-1}$:
- Step 1: Compute determinant: $\det(\mathbf{A}) = 1 \cdot 4 - 2 \cdot 3 = -2$.
- Step 2: Use formula for 2x2 inverse:
  $$
  \mathbf{A}^{-1} = \frac{1}{\det(\mathbf{A})} \begin{bmatrix} 4 & -2 \\ -3 & 1 \end{bmatrix} = \begin{bmatrix} -2 & 1 \\ 1.5 & -0.5 \end{bmatrix}
  $$

**Why Necessary**: Solves linear systems and supports analytical solutions in ML.

**ML Use-Case**: In linear regression, the normal equation uses the inverse: $\mathbf{w} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$.

```python
# Matrix Inverse
import numpy as np

# Define matrix
A = np.array([[1, 2], [3, 4]])
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)

print("Matrix Inverse:\n", A_inv)
print("A * A_inv (Identity):\n", I)
```

```python
# Example 2: Linear Regression (Normal Equation)
import numpy as np

# Simulate data: X (3 samples, 2 features), y (3 targets)
X = np.array([[1, 2], [3, 4], [5, 6]])
y = np.array([1, 2, 3])

# Normal equation: w = (X^T X)^(-1) X^T y
X_T = X.T
w = np.linalg.inv(X_T @ X) @ X_T @ y
print("Linear Regression Weights:\n", w)
```
