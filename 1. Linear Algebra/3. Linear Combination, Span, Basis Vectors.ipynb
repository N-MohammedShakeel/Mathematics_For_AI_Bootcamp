# Linear Combinations, Span, and Basis Vectors in Linear Algebra for AI/ML

## Introduction and Relevance
**Definition**: Linear combinations, span, and basis vectors are foundational concepts in linear algebra that describe how vectors generate other vectors or subspaces. These concepts are critical in AI/ML for representing data, modeling transformations, and reducing dimensionality.

**Mathematical Definitions**:
- **Linear Combination**: A vector $\mathbf{u}$ formed as:
  $$
  \mathbf{u} = c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \dots + c_k \mathbf{v}_k
  $$
  where $\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k$ are vectors in a vector space $V$, and $c_1, c_2, \dots, c_k \in \mathbb{R}$ are scalar coefficients.
- **Span**: The set of all possible linear combinations of a set of vectors $\{\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k\}$:
  $$
  \text{span}(\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k) = \{ c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \dots + c_k \mathbf{v}_k \mid c_i \in \mathbb{R} \}
  $$
  The span is a subspace of $V$.
- **Basis Vectors**: A set $\{\mathbf{b}_1, \mathbf{b}_2, \dots, \mathbf{b}_n\}$ is a basis for $V$ if:
  - The vectors are **linearly independent** (no vector is a linear combination of the others).
  - The vectors **span** $V$ (every vector in $V$ is a linear combination of the basis vectors).
  - The number of basis vectors equals the dimension of $V$.

**Examples**:
- **Linear Combination**: For $\mathbf{v}_1 = [1, 0]$, $\mathbf{v}_2 = [0, 1]$, the vector $[3, 2] = 3 \mathbf{v}_1 + 2 \mathbf{v}_2$ is a linear combination.
- **Span**: The span of $\{[1, 0], [0, 1]\}$ in $\mathbb{R}^2$ is all of $\mathbb{R}^2$.
- **Basis**: The standard basis for $\mathbb{R}^2$ is $\{[1, 0], [0, 1]\}$, which is linearly independent and spans $\mathbb{R}^2$.

**Geometric Interpretation**:
- A linear combination scales and adds vectors, producing a new vector in the direction defined by the input vectors.
- The span of vectors forms a subspace (e.g., a line, plane, or higher-dimensional space) depending on the number of linearly independent vectors.
- Basis vectors act as a coordinate system, allowing unique representation of any vector in the space.

**Relevance in Mathematics**:
- **Vector Space Structure**: Define how vectors generate subspaces through combinations.
- **Coordinate Systems**: Represent vectors as coordinates relative to a basis.
- **Dimensionality Analysis**: Determine the dimension of a space via the number of basis vectors.
- **Transformations**: Express linear transformations as combinations of basis vectors.

**Relevance in AI/ML**:
- **Data Representation**: Express data points as linear combinations of features or basis vectors.
- **Subspaces**: The span defines the effective feature space or solution space.
- **Dimensionality Reduction**: Basis vectors (e.g., principal components) reduce data dimensionality.
- **Efficient Computations**: Linearly independent bases simplify matrix operations in ML algorithms.

**Applications Overview**:
- **PCA**: Basis vectors (principal components) span a reduced-dimensional subspace.
- **Feature Engineering**: Linear combinations create new feature representations.
- **Neural Networks**: Outputs of linear layers are linear combinations of inputs.
- **Sparse Coding**: Basis vectors provide sparse signal representations.
- **Clustering**: Eigenvectors as basis vectors define clusters in spectral methods.
- **Recommendation Systems**: Linear combinations of latent factors represent user/item profiles.

### 2.1 Linear Combinations
**Definition**: A linear combination of vectors $\{\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k\}$ is:
  $$
  \mathbf{u} = c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \dots + c_k \mathbf{v}_k
  $$
where $c_i$ are scalars.

**Explanation**: Linear combinations allow vectors to be constructed by scaling and adding other vectors. In ML, data points (e.g., feature vectors) are often linear combinations of basis vectors or features, enabling flexible data representation.

**Derivation**: Consider vectors $\mathbf{v}_1, \mathbf{v}_2$ in $\mathbb{R}^2$. A linear combination $c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2$ generates a vector by scaling $\mathbf{v}_1$ by $c_1$, $\mathbf{v}_2$ by $c_2$, and adding the results. Geometrically, this traces a path in the direction of the vectors, constrained by their span.

**Properties**:
- **Closure**: The result remains in the vector space.
- **Flexibility**: Any vector in the span can be expressed as a linear combination.
- **Uniqueness**: If the vectors are linearly independent, the coefficients $c_i$ are unique.

**Why Necessary**:
- **Data Representation**: Represents complex data (e.g., images, text) as combinations of simpler components.
- **Model Transformations**: Linear layers in neural networks compute linear combinations of inputs.
- **Feature Engineering**: Creates new features as combinations of existing ones.
- **Generalization**: Underpins the construction of subspaces in ML algorithms.

**Example**: For $\mathbf{v}_1 = [1, 0]$, $\mathbf{v}_2 = [0, 1]$, the vector $[3, 2] = 3 \mathbf{v}_1 + 2 \mathbf{v}_2$.

**ML Use-Case**: In neural networks, a neuron’s output is a linear combination of input features weighted by learned parameters.

```python
# Linear Combination Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
v1 = np.array([1, 0])
v2 = np.array([0, 1])
c1, c2 = 3, 2
u = c1 * v1 + c2 * v2  # Linear combination

# Visualize
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], color='blue', scale=1, scale_units='xy', angles='xy', label='v₁ = [1, 0]')
plt.quiver(0, 0, v2[0], v2[1], color='green', scale=1, scale_units='xy', angles='xy', label='v₂ = [0, 1]')
plt.quiver(0, 0, c1*v1[0], c1*v1[1], color='blue', alpha=0.5, scale=1, scale_units='xy', angles='xy', label=f'{c1}v₁')
plt.quiver(c1*v1[0], c1*v1[1], c2*v2[0], c2*v2[1], color='green', alpha=0.5, scale=1, scale_units='xy', angles='xy', label=f'{c2}v₂')
plt.quiver(0, 0, u[0], u[1], color='red', scale=1, scale_units='xy', angles='xy', label=f'u = {c1}v₁ + {c2}v₂ = [{u[0]}, {u[1]}]')
plt.xlim(-1, 4)
plt.ylim(-1, 3)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Linear Combination in $\mathbb{R}^2$')
plt.legend()
plt.show()

print(f'Linear combination: {c1} * {v1} + {c2} * {v2} = {u}')
```

**Visualization Explanation**: The plot shows vectors $\mathbf{v}_1 = [1, 0]$ and $\mathbf{v}_2 = [0, 1]$, their scaled versions $3 \mathbf{v}_1$ and $2 \mathbf{v}_2$, and the resulting linear combination $\mathbf{u} = [3, 2]$. The red arrow represents $\mathbf{u}$, illustrating how it is constructed by adding the scaled vectors, emphasizing the geometric interpretation of linear combinations.

### 2.2 Span
**Definition**: The span of vectors $\{\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k\}$ is the subspace of all their linear combinations:
  $$
  \text{span}(\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k) = \{ c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \dots + c_k \mathbf{v}_k \mid c_i \in \mathbb{R} \}
  $$

**Explanation**: The span represents the full extent of the subspace generated by the vectors, such as a line (1D), plane (2D), or higher-dimensional space, depending on the number of linearly independent vectors. In ML, the span defines the feature space or solution space for data or models.

**Derivation**: The span is the column space of the matrix $\mathbf{V} = [\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k]$. The dimension of the span is the rank of $\mathbf{V}$, determined by the number of linearly independent vectors. For example, in $\mathbb{R}^2$, two linearly independent vectors span the entire plane, while a single vector spans a line.

**Properties**:
- **Subspace**: The span is closed under addition and scalar multiplication.
- **Redundancy**: Adding linearly dependent vectors does not increase the span.
- **Dimension**: Equals the rank of the matrix formed by the vectors.

**Why Necessary**:
- **Feature Space Definition**: Determines the subspace that data points occupy.
- **Dimensionality Reduction**: Identifies the effective dimensionality of a dataset (e.g., in PCA).
- **Transformation Analysis**: Defines the range of linear transformations in ML models.
- **Data Coverage**: Ensures a set of vectors can represent all points in a space.

**Example**: The span of $\{[1, 0], [0, 1]\}$ is $\mathbb{R}^2$, while the span of $\{[1, 0]\}$ is a line along the x-axis.

**ML Use-Case**: In PCA, the span of principal components defines the reduced-dimensional subspace capturing maximum data variance.

```python
# ---------------------------
# Span Visualization Example
# ---------------------------

import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# 1. Define the basis vectors
# ---------------------------

# First vector v1 = [1, 0] (unit vector along x-axis)
v1 = np.array([1, 0])

# Second vector v2 = [0, 1] (unit vector along y-axis)
v2 = np.array([0, 1])

# Stack the vectors as columns in a matrix V
# Shape of V will be (2, 2)
# V = [[1, 0],
#      [0, 1]]
V = np.array([v1, v2]).T

# ---------------------------
# 2. Generate points in the span
# ---------------------------

# Choose coefficients for linear combinations
# These represent all possible scalars c1 and c2 that multiply v1 and v2
c1 = np.linspace(-2, 2, 20)  # 20 values between -2 and 2 for v1's coefficient
c2 = np.linspace(-2, 2, 20)  # 20 values between -2 and 2 for v2's coefficient

# Create a 2D grid of all combinations of c1 and c2
# C1 and C2 are 20x20 matrices representing all coefficient pairs
C1, C2 = np.meshgrid(c1, c2)

# Generate points in the span:
# Each point = c1*v1 + c2*v2
# This is the definition of "span" → all possible linear combinations of vectors
# The shape of `points` will be (20, 20, 2) because each point has 2 coordinates (x, y)
points = C1[:, :, np.newaxis] * v1 + C2[:, :, np.newaxis] * v2

# ---------------------------
# 3. Visualization of span
# ---------------------------

plt.figure(figsize=(6, 6))  # Square plot for symmetry

# Scatter plot of all span points (20x20 = 400 points)
# Points are colored blue with alpha=0.1 (transparent, so density can be seen)
plt.scatter(points[:, :, 0], points[:, :, 1], 
            c='blue', alpha=0.1, label='Span of {v₁, v₂}')

# Draw vector v1 as an arrow from origin
plt.quiver(0, 0, v1[0], v1[1], 
           color='red', scale=1, scale_units='xy', angles='xy', 
           label='v₁ = [1, 0]')

# Draw vector v2 as an arrow from origin
plt.quiver(0, 0, v2[0], v2[1], 
           color='green', scale=1, scale_units='xy', angles='xy', 
           label='v₂ = [0, 1]')

# Set limits for the plot (so span is clearly visible)
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# Add gridlines
plt.grid(True)

# Draw x-axis and y-axis for reference
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Add title
plt.title('Span of $\{[1, 0], [0, 1]\}$ (All of $\mathbb{R}^2$)')

# Show legend with v1, v2, and span
plt.legend()

# Display the figure
plt.show()

# ---------------------------
# 4. Rank check
# ---------------------------

# The rank of the matrix formed by v1 and v2 tells us
# how many independent directions (dimensions) they span
rank = np.linalg.matrix_rank(V)

# Print result
# If rank = 2 → the span covers all of R^2
print(f'Rank of span: {rank} (spans $\mathbb{{R}}^{rank}$ if rank = 2)')
```

**Visualization Explanation**: The plot shows the span of $\{[1, 0], [0, 1]\}$, which covers all of $\mathbb{R}^2$ as a plane (blue points). The vectors $\mathbf{v}_1$ and $\mathbf{v}_2$ are shown as arrows, and the rank confirms that they span a 2D space, illustrating that all points in $\mathbb{R}^2$ can be reached via their linear combinations.

### 2.3 Basis Vectors
**Definition**: A basis for a vector space $V$ is a set of linearly independent vectors $\{\mathbf{b}_1, \mathbf{b}_2, \dots, \mathbf{b}_n\}$ that spans $V$. The number $n$ is the dimension of $V$.

**Explanation**: Basis vectors provide a coordinate system for the space, allowing every vector to be uniquely represented as a linear combination. In ML, bases (e.g., principal components) simplify data representation and transformations.

**Derivation**: A set of vectors is a basis if the matrix formed by the vectors has full rank (equal to the dimension of $V$). For example, in $\mathbb{R}^2$, the standard basis $\{[1, 0], [0, 1]\}$ has rank 2, spans $\mathbb{R}^2$, and is linearly independent.

**Properties**:
- **Linear Independence**: No basis vector is a linear combination of others.
- **Spanning**: The basis spans the entire space.
- **Uniqueness of Representation**: Each vector has unique coordinates relative to the basis.
- **Orthonormal Basis**: If vectors are orthogonal and normalized, computations (e.g., projections) are simplified.

**Why Necessary**:
- **Coordinate System**: Provides a framework for representing data points or transformations.
- **Dimensionality Reduction**: Subsets of basis vectors (e.g., PCA) reduce data dimensionality.
- **Efficient Computations**: Orthonormal bases simplify matrix operations in ML.
- **Model Interpretability**: Basis vectors reveal key directions in data (e.g., principal components).

**Example**: The standard basis for $\mathbb{R}^3$ is $\{[1, 0, 0], [0, 1, 0], [0, 0, 1]\}$.

**ML Use-Case**: In PCA, principal components form an orthonormal basis for a reduced-dimensional subspace.

```python
# Basis Vectors Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define basis vectors
b1 = np.array([1, 0])
b2 = np.array([0, 1])
B = np.array([b1, b2]).T  # Matrix with basis vectors as columns

# Generate a test vector
u = np.array([3, 2])

# Visualize
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, b1[0], b1[1], color='blue', scale=1, scale_units='xy', angles='xy', label='b₁ = [1, 0]')
plt.quiver(0, 0, b2[0], b2[1], color='green', scale=1, scale_units='xy', angles='xy', label='b₂ = [0, 1]')
plt.quiver(0, 0, u[0], u[1], color='red', scale=1, scale_units='xy', angles='xy', label=f'u = [3, 2]')
plt.xlim(-1, 4)
plt.ylim(-1, 3)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Basis Vectors and Vector Representation in $\mathbb{R}^2$')
plt.legend()
plt.show()

rank = np.linalg.matrix_rank(B)
print(f'Rank of basis: {rank} (basis for $\mathbb{{R}}^{rank}$ if rank = 2)')
print(f'Coordinates of u = [3, 2] in basis {{b₁, b₂}}: [{u[0]}, {u[1]}]')
```

**Visualization Explanation**: The plot shows the standard basis $\{[1, 0], [0, 1]\}$ for $\mathbb{R}^2$ and a vector $\mathbf{u} = [3, 2]$. The basis vectors form a coordinate system, and $\mathbf{u}$ is represented as $3 \mathbf{b}_1 + 2 \mathbf{b}_2$, demonstrating how basis vectors allow unique vector representation.

### 2.4 Linear Independence and Dependence
**Definition**: A set of vectors is **linearly independent** if no vector is a linear combination of the others; otherwise, it is **linearly dependent**. Mathematically, $\{\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k\}$ is linearly independent if:
  $$
  c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \dots + c_k \mathbf{v}_k = 0 \implies c_1 = c_2 = \dots = c_k = 0
  $$

**Explanation**: Linear independence ensures that each vector contributes unique information, critical for forming a basis. Linear dependence indicates redundancy, which can complicate ML models.

**Derivation**: Linear independence is tested by forming the matrix $\mathbf{V} = [\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k]$ and checking its rank. If $\text{rank}(\mathbf{V}) = k$, the vectors are independent; if less, some are dependent.

**Why Necessary**:
- **Basis Formation**: Only linearly independent vectors can form a basis.
- **Feature Selection**: Avoids multicollinearity in regression models.
- **Sparse Representations**: Ensures unique signal reconstruction in compressed sensing.
- **Model Efficiency**: Reduces redundancy in feature or parameter spaces.

**Example**: $\{[1, 0], [0, 1], [2, 0]\}$ is linearly dependent because $[2, 0] = 2 [1, 0]$.

**ML Use-Case**: In feature selection, linearly independent features prevent overfitting and improve model stability.

```python
# Linear Independence Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define vectors (including a dependent one)
v1 = np.array([1, 0])
v2 = np.array([0, 1])
v3 = np.array([2, 0])  # Dependent: v3 = 2 * v1
V = np.array([v1, v2, v3]).T

# Visualize
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], color='blue', scale=1, scale_units='xy', angles='xy', label='v₁ = [1, 0]')
plt.quiver(0, 0, v2[0], v2[1], color='green', scale=1, scale_units='xy', angles='xy', label='v₂ = [0, 1]')
plt.quiver(0, 0, v3[0], v3[1], color='red', scale=1, scale_units='xy', angles='xy', label='v₃ = [2, 0] (dependent)')
plt.xlim(-1, 3)
plt.ylim(-1, 2)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Linear Dependence: $\{v₁, v₂, v₃\}$ (v₃ = 2v₁)')
plt.legend()
plt.show()

rank = np.linalg.matrix_rank(V)
print(f'Number of linearly independent vectors: {rank} (out of {V.shape[1]})')
```

**Visualization Explanation**: The plot shows vectors $\mathbf{v}_1 = [1, 0]$, $\mathbf{v}_2 = [0, 1]$, and $\mathbf{v}_3 = [2, 0]$, where $\mathbf{v}_3$ lies along the same direction as $\mathbf{v}_1$, indicating linear dependence. The rank confirms only two vectors are independent, highlighting redundancy.

## Practical ML Examples
Below are coded examples demonstrating linear combinations, span, and basis vectors in ML contexts.

```python
# Example 1: PCA (Basis Vectors and Span)
import numpy as np
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(0)
data = np.random.multivariate_normal([0, 0], [[2, 1], [1, 2]], 100)

# Center data
centered_data = data - np.mean(data, axis=0)

# Compute covariance matrix and eigenvectors
cov_matrix = np.cov(centered_data.T)
eig_vals, eig_vecs = np.linalg.eigh(cov_matrix)

# Sort eigenvectors by eigenvalues (descending)
idx = np.argsort(eig_vals)[::-1]
eig_vecs = eig_vecs[:, idx]

# Project data onto principal components
projected_data = centered_data @ eig_vecs

# Visualize
plt.figure(figsize=(6, 6))
plt.scatter(centered_data[:, 0], centered_data[:, 1], alpha=0.5, label='Centered Data')
plt.quiver(0, 0, eig_vecs[0, 0], eig_vecs[1, 0], color='red', scale=5, label='PC1')
plt.quiver(0, 0, eig_vecs[0, 1], eig_vecs[1, 1], color='green', scale=5, label='PC2')
plt.scatter(projected_data[:, 0], projected_data[:, 1], alpha=0.3, color='purple', label='Projected Data')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('PCA: Basis Vectors (Principal Components)')
plt.legend()
plt.show()

print('Principal components (basis vectors):\n', eig_vecs)
print('Projected data (first 5 points):\n', projected_data[:5])
```

**Example Explanation**: This code performs PCA on simulated 2D data, computing principal components (basis vectors) via the covariance matrix’s eigenvectors. The plot shows the data, principal components (red and green arrows), and projected data, illustrating how the basis spans the subspace capturing maximum variance.
