# Vectors in Linear Algebra for AI/ML

## Introduction and Relevance
**Definition**: A vector is a mathematical object representing a quantity with both magnitude and direction, expressed as an ordered list of numbers in an n-dimensional space ($\mathbb{R}^n$). In AI/ML, vectors are one-dimensional arrays (e.g., $\mathbf{v} = [v_1, v_2, \dots, v_n]$) used to represent data points, features, or model parameters.

**Geometric Interpretation**: In 2D/3D space, a vector is an arrow from the origin to a point, with components defining its position (e.g., $\mathbf{v} = [3, 4]$ points to $(3, 4)$ in 2D).

**Algebraic Interpretation**: A vector is a list of numbers manipulated via operations like addition or scalar multiplication.

**Examples**:
- 2D vector: $\mathbf{v} = [1, 2]$
- 3D vector: $\mathbf{u} = [3, -1, 4]$
- High-dimensional vector: A feature vector in ML, e.g., $\mathbf{x} = [0.5, 1.2, 3.7, \dots]$ for pixel intensities or word embeddings.

**Relevance in Mathematics**:
- Vectors form the building blocks of vector spaces, enabling geometric analysis, linear combinations, and generalizations to higher dimensions.
- Operations define properties like linearity, orthogonality, and norms, underpinning concepts like eigenvalues and SVD.

**Relevance in AI/ML**:
- **Data Representation**: Encode inputs, features, or embeddings (e.g., images, text).
- **Model Parameters**: Represent weights, biases, or gradients in models.
- **Computations**: Enable efficient operations like similarity measures and optimization.
- **High-Dimensional Spaces**: Handle complex datasets in high-dimensional feature spaces.

**Applications Overview**:
- **Data Preprocessing**: Normalizing or combining feature vectors.
- **Neural Networks**: Computing weighted sums and updating parameters.
- **NLP**: Measuring similarity between word embeddings.
- **Computer Vision**: Representing 3D coordinates or pixel intensities.
- **Optimization**: Updating model parameters via gradient descent.
- **Recommendation Systems**: Computing user-item similarities.

### 2.1 Vector Representation
**Formula**: A vector $\mathbf{v} \in \mathbb{R}^n$ is represented as $\mathbf{v} = [v_1, v_2, \dots, v_n]$, where each $v_i$ is a scalar (real number).

**Explanation**: Vectors encode data in a structured format, representing points or directions in space. In ML, they represent data points (e.g., a row in a dataset) or parameters (e.g., neural network weights).

**Example**: A 2D vector $\mathbf{v} = [1, 2]$ represents a point at $(1, 2)$. In ML, a feature vector like $[0.5, 1.2, 3.7]$ might represent pixel intensities.

**Why Necessary**: Provides a standardized way to store and manipulate multidimensional data.

```python
# Vector Representation
import numpy as np

# Define a 2D vector
v = np.array([1, 2])
print("2D Vector:", v)

# Define a high-dimensional feature vector (e.g., ML data point)
feature_vector = np.array([0.5, 1.2, 3.7, 2.1])
print("Feature Vector:", feature_vector)
```

### 2.2 Vector Addition
**Formula**: For vectors $\mathbf{u} = [u_1, u_2, \dots, u_n]$ and $\mathbf{v} = [v_1, v_2, \dots, v_n]$, the sum is:
$$
\mathbf{u} + \mathbf{v} = [u_1 + v_1, u_2 + v_2, \dots, u_n + v_n]
$$

**Explanation**: Vector addition combines vectors element-wise, representing the cumulative effect of two directions. Geometrically, it follows the parallelogram law.

**Example**: Given $\mathbf{u} = [1, 2]$ and $\mathbf{v} = [3, 4]$, compute $\mathbf{u} + \mathbf{v}$:
- Step 1: Add components:
  $$
  \mathbf{u} + \mathbf{v} = [1 + 3, 2 + 4] = [4, 6]
  $$

**Why Necessary**: Combines information from multiple sources, used in ML for aggregating features or updating parameters (e.g., gradient descent).

**ML Use-Case**: In gradient descent, weights are updated as $\mathbf{w} \leftarrow \mathbf{w} - \eta \nabla L$, where subtraction is addition of the negative vector.

```python
# Vector Addition Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
u = np.array([1, 2])
v = np.array([3, 4])
sum_uv = u + v

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u = [1, 2]')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='v = [3, 4]')
plt.quiver(0, 0, sum_uv[0], sum_uv[1], angles='xy', scale_units='xy', scale=1, color='green', label='u + v = [4, 6]')
plt.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5)

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Vector Addition (Parallelogram Law)')
plt.legend()
plt.show()

print("Vector Addition Result:", sum_uv)
```

### 2.3 Scalar Multiplication
**Formula**: For a vector $\mathbf{v} = [v_1, v_2, \dots, v_n]$ and scalar $c$, the result is:
$$
c \mathbf{v} = [c v_1, c v_2, \dots, c v_n]
$$

**Explanation**: Scales the vectorâ€™s magnitude by $c$, stretching or shrinking it while preserving or reversing direction (if $c < 0$).

**Example**: Given $\mathbf{v} = [1, 2]$ and $c = 3$, compute $c \mathbf{v}$:
- Step 1: Multiply components:
  $$
  c \mathbf{v} = [3 \cdot 1, 3 \cdot 2] = [3, 6]
  $$

**Why Necessary**: Adjusts vector magnitude, critical for normalization or scaling in ML (e.g., feature scaling, gradient steps).

**ML Use-Case**: In gradient descent, the learning rate $\eta$ scales the gradient: $\mathbf{w} \leftarrow \mathbf{w} - \eta \nabla L$.

```python
# Scalar Multiplication Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define vector and scalar
v = np.array([1, 2])
c = 3
scaled_v = c * v

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v = [1, 2]')
plt.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color='purple', label='3 * v = [3, 6]')

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Scalar Multiplication')
plt.legend()
plt.show()

print("Scalar Multiplication Result:", scaled_v)
```

### 2.4 Dot Product
**Formula**: For vectors $\mathbf{u} = [u_1, u_2, \dots, u_n]$ and $\mathbf{v} = [v_1, v_2, \dots, v_n]$, the dot product is:
$$
\mathbf{u} \cdot \mathbf{v} = u_1 v_1 + u_2 v_2 + \dots + u_n v_n
$$
Geometrically: $\mathbf{u} \cdot \mathbf{v} = |\mathbf{u}| |\mathbf{v}| \cos(\theta)$.

**Explanation**: Measures similarity or projection of one vector onto another. A positive/negative/zero value indicates acute/obtuse/orthogonal angles.

**Example**: Given $\mathbf{u} = [1, 2]$ and $\mathbf{v} = [3, 4]$, compute $\mathbf{u} \cdot \mathbf{v}$:
- Step 1: Multiply and sum components:
  $$
  \mathbf{u} \cdot \mathbf{v} = 1 \cdot 3 + 2 \cdot 4 = 3 + 8 = 11
  $$
- Verify geometrically:
  $$
  |\mathbf{u}| = \sqrt{1^2 + 2^2} = \sqrt{5}, \quad |\mathbf{v}| = \sqrt{3^2 + 4^2} = 5, \quad \cos(\theta) = \frac{11}{\sqrt{5} \cdot 5}
  $$

**Why Necessary**: Quantifies similarity, critical for ML tasks like classification or clustering.

**ML Use-Case**: In neural networks, the dot product computes weighted sums: $\mathbf{w} \cdot \mathbf{x}$. In NLP, itâ€™s used for cosine similarity.

```python
# Dot Product Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
u = np.array([1, 2])
v = np.array([3, 4])

# Compute dot product
dot_product = np.dot(u, v)

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u = [1, 2]')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='v = [3, 4]')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title(f'Dot Product Visualization (Value = {dot_product})')
plt.legend()
plt.show()

print("Dot Product:", dot_product)
```

### 2.5 Cross Product
**Formula**: For 3D vectors $\mathbf{u} = [u_1, u_2, u_3]$ and $\mathbf{v} = [v_1, v_2, v_3]$, the cross product is:
$$
\mathbf{u} \times \mathbf{v} = [u_2 v_3 - u_3 v_2, u_3 v_1 - u_1 v_3, u_1 v_2 - u_2 v_1]
$$

**Explanation**: Produces a vector perpendicular to both inputs, with magnitude equal to the area of the parallelogram they span.

**Example**: Given $\mathbf{u} = [1, 0, 0]$ and $\mathbf{v} = [0, 1, 0]$, compute $\mathbf{u} \times \mathbf{v}$:
- Step 1: Apply the formula:
  $$
  \mathbf{u} \times \mathbf{v} = [(0 \cdot 0 - 0 \cdot 1), (0 \cdot 0 - 1 \cdot 0), (1 \cdot 1 - 0 \cdot 0)] = [0, 0, 1]
  $$

**Why Necessary**: Critical for 3D applications like computer vision or robotics to compute normals or torque.

**ML Use-Case**: In computer vision, cross products compute surface normals for 3D reconstruction.

```python
# Cross Product Visualization (3D)
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define vectors
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
cross_product = np.cross(u, v)

# Plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='blue', label='u = [1, 0, 0]')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='red', label='v = [0, 1, 0]')
ax.quiver(0, 0, 0, cross_product[0], cross_product[1], cross_product[2], color='green', label='u Ã— v = [0, 0, 1]')

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Cross Product Visualization')
plt.legend()
plt.show()

print("Cross Product:", cross_product)
```

### 2.6 Norm and Cosine Similarity
**Formula**:
- Norm: $|\mathbf{v}| = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}$
- Cosine Similarity: $\cos(\theta) = \frac{\mathbf{u} \cdot \mathbf{v}}{|\mathbf{u}| |\mathbf{v}|}$

**Explanation**: The norm measures a vectorâ€™s magnitude, while cosine similarity quantifies the angle between vectors, indicating their similarity.

**Example**: For $\mathbf{u} = [1, 2]$ and $\mathbf{v} = [3, 4]$:
- Norm of $\mathbf{u}$:
  $$
  |\mathbf{u}| = \sqrt{1^2 + 2^2} = \sqrt{5}
  $$
- Cosine similarity:
  $$
  \cos(\theta) = \frac{1 \cdot 3 + 2 \cdot 4}{\sqrt{5} \cdot \sqrt{3^2 + 4^2}} = \frac{11}{\sqrt{5} \cdot 5}
  $$

**Why Necessary**: Norms normalize data; cosine similarity measures alignment, crucial for ML tasks like NLP.

**ML Use-Case**: Cosine similarity in NLP compares word embeddings (e.g., â€œkingâ€ vs. â€œqueenâ€).

```python
# Norm and Cosine Similarity
import numpy as np

# Define vectors
u = np.array([1, 2])
v = np.array([3, 4])

# Compute norm and cosine similarity
norm_u = np.linalg.norm(u)
cos_sim = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

print("Norm of u:", norm_u)
print("Cosine Similarity:", cos_sim)
```

## Practical ML Examples
Below are coded examples demonstrating vector operations in ML contexts.

```python
# Example 1: Cosine Similarity for Word Embeddings
import numpy as np

# Simulated word embeddings
king = np.array([0.8, 0.1, 0.3, 0.2])
queen = np.array([0.7, 0.2, 0.4, 0.1])

# Cosine similarity
cos_sim = np.dot(king, queen) / (np.linalg.norm(king) * np.linalg.norm(queen))
print("Cosine Similarity between 'king' and 'queen':", cos_sim)
```

```python
# Example 2: Gradient Descent Update
import numpy as np

# Simulate weights and gradient
weights = np.array([1.0, 2.0, 3.0])
gradient = np.array([0.1, 0.2, 0.3])
learning_rate = 0.01

# Update weights
updated_weights = weights - learning_rate * gradient
print("Original Weights:", weights)
print("Updated Weights:", updated_weights)
```
