# Norms and Distances in Linear Algebra for AI/ML

## Introduction and Relevance
**Definition**: Norms and distances are mathematical tools used to measure the size of vectors and the separation between points in a vector space, respectively. They are critical in AI/ML for quantifying similarity, error, and data relationships in high-dimensional spaces.

**Mathematical Definitions**:
- **Norm**: A function $\|\cdot\|: V \to \mathbb{R}$ (where $V$ is a vector space, typically $\mathbb{R}^n$) assigns a non-negative scalar to a vector, representing its "length" or "magnitude".
  - Properties:
    - Non-negativity: $\|\mathbf{v}\| \geq 0$, and $\|\mathbf{v}\| = 0$ if and only if $\mathbf{v} = \mathbf{0}$.
    - Homogeneity: $\|c \mathbf{v}\| = |c| \|\mathbf{v}\|$ for scalar $c \in \mathbb{R}$.
    - Triangle Inequality: $\|\mathbf{u} + \mathbf{v}\| \leq \|\mathbf{u}\| + \|\mathbf{v}\|$.
- **Distance**: A function $d: V \times V \to \mathbb{R}$ measures separation, typically $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|$.
  - Properties:
    - Non-negativity: $d(\mathbf{u}, \mathbf{v}) \geq 0$, and $d(\mathbf{u}, \mathbf{v}) = 0$ if and only if $\mathbf{u} = \mathbf{v}$.
    - Symmetry: $d(\mathbf{u}, \mathbf{v}) = d(\mathbf{v}, \mathbf{u})$.
    - Triangle Inequality: $d(\mathbf{u}, \mathbf{w}) \leq d(\mathbf{u}, \mathbf{v}) + d(\mathbf{v}, \mathbf{w})$.

**Examples**:
- **Euclidean Norm**: $\|\mathbf{v}\|_2 = \sqrt{v_1^2 + \dots + v_n^2}$.
- **Euclidean Distance**: $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|_2$.
- **ML Context**: Norms measure feature vector or parameter magnitudes; distances quantify similarity (e.g., in clustering).

**Geometric Interpretation**:
- Norms represent the "length" of a vector from the origin.
- Distances measure the straight-line or other separation between points.

**Relevance in Mathematics**:
- **Vector Measurement**: Quantify vector magnitudes consistently.
- **Metric Spaces**: Define geometric structures for data relationships.
- **Optimization**: Provide objective functions (e.g., minimizing error norms).
- **Geometric Insights**: Measure distances and angles in vector spaces.

**Relevance in AI/ML**:
- **Similarity**: Quantify data point proximity for clustering/classification.
- **Regularization**: Penalize model parameters to prevent overfitting.
- **Error Measurement**: Quantify residuals in regression or optimization.
- **Geometric Algorithms**: Support PCA, k-NN, and clustering.

**Applications Overview**:
- **k-Nearest Neighbors**: Distances identify closest points.
- **Clustering**: Euclidean distance minimizes within-cluster variance.
- **Regularization**: L1/L2 norms control model complexity.
- **PCA**: Norms normalize eigenvectors; distances measure variance.
- **Text/NLP**: Cosine distance measures embedding similarity.
- **Anomaly Detection**: Norms identify outliers.
- **Neural Networks**: Norms guide weight optimization.

## Core Concepts and Derivations
This section covers types of norms, distances, norm-induced metrics, and norms with orthogonality, with detailed derivations, explanations, and visualizations.

### 2.1 Types of Norms
**Definition**: Common norms in ML include:
- **L2 Norm (Euclidean)**: $\|\mathbf{v}\|_2 = \sqrt{\sum_{i=1}^n v_i^2}$.
- **L1 Norm (Manhattan)**: $\|\mathbf{v}\|_1 = \sum_{i=1}^n |v_i|$.
- **Infinity Norm (Max)**: $\|\mathbf{v}\|_\infty = \max_i |v_i|$.
- **Lp Norm**: $\|\mathbf{v}\|_p = \left( \sum_{i=1}^n |v_i|^p \right)^{1/p}$, $p \geq 1$.

**Explanation**: Norms measure vector magnitude, with L1 promoting sparsity, L2 emphasizing large components, and L-infinity focusing on the largest component.

**Derivation**: The L2 norm derives from the Euclidean distance in $\mathbb{R}^n$, linked to the dot product: $\|\mathbf{v}\|_2 = \sqrt{\mathbf{v} \cdot \mathbf{v}}$. The L1 norm sums absolute values, useful for sparse solutions. The Lp norm generalizes these, satisfying norm properties via Minkowski’s inequality.

**Properties**:
- **Equivalence**: All norms are equivalent in finite-dimensional spaces but differ in sensitivity.
- **Convexity**: Norms are convex, aiding optimization.

**Why Necessary**:
- **Regularization**: L1/L2 norms penalize parameters.
- **Error Measurement**: Quantify residuals or gradients.
- **Feature Scaling**: Normalize data for consistent scales.
- **Optimization**: Guide step sizes in gradient descent.

**Example**: For $\mathbf{v} = [3, 4]$, $\|\mathbf{v}\|_2 = 5$, $\|\mathbf{v}\|_1 = 7$, $\|\mathbf{v}\|_\infty = 4$.

**ML Use-Case**: L2 norm in ridge regression, L1 in Lasso for sparsity.

```python
# Norm Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define vector
v = np.array([3, 4])

# Compute norms
l2_norm = np.linalg.norm(v)
l1_norm = np.linalg.norm(v, 1)
inf_norm = np.linalg.norm(v, np.inf)

# Visualize
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v = [3, 4]')
plt.plot([0, v[0]], [v[1], v[1]], 'r--', label='L1 path (|x| + |y| = 7)')
plt.plot([0, v[0]], [0, 0], 'g--', label='L-infinity path (max = 4)')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Vector Norms Visualization')
plt.legend()
plt.show()

print(f'L2 norm: {l2_norm:.2f}')
print(f'L1 norm: {l1_norm:.2f}')
print(f'Infinity norm: {inf_norm:.2f}')
```

**Visualization Explanation**: The plot shows vector $\mathbf{v} = [3, 4]$, with dashed lines illustrating L1 (Manhattan) and L-infinity paths, contrasting with the L2 (Euclidean) straight-line distance. This highlights how different norms measure magnitude differently.

### 2.2 Types of Distances
**Definition**: Common distances include:
- **Euclidean Distance**: $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|_2$.
- **Manhattan Distance**: $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|_1$.
- **Chebyshev Distance**: $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|_\infty$.
- **Cosine Distance**: $d(\mathbf{u}, \mathbf{v}) = 1 - \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\|_2 \|\mathbf{v}\|_2}$.

**Explanation**: Distances quantify separation, with Euclidean focusing on straight-line distance, Manhattan on grid-like paths, Chebyshev on maximum differences, and cosine on angular separation.

**Derivation**: Euclidean distance derives from the L2 norm, linked to the Pythagorean theorem. Cosine distance uses the dot product to measure angle, independent of magnitude.

**Properties**:
- **Metric Space**: Distances define a metric space, ensuring consistent measurements.
- **Sensitivity**: Euclidean is magnitude-sensitive; cosine focuses on orientation.

**Why Necessary**:
- **Similarity**: Essential for clustering and classification (e.g., k-NN).
- **Error Evaluation**: Measure residuals in regression.
- **Data Relationships**: Quantify proximity in high-dimensional spaces.

**Example**: For $\mathbf{u} = [1, 2]$, $\mathbf{v} = [3, 4]$, Euclidean distance is $\sqrt{8} \approx 2.828$, Manhattan is 4, cosine distance is $1 - \cos(\theta)$.

**ML Use-Case**: Cosine distance in NLP for document similarity.

```python
# Distance Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
u = np.array([1, 2])
v = np.array([3, 4])

# Compute distances
euclidean_dist = np.linalg.norm(u - v)
manhattan_dist = np.linalg.norm(u - v, 1)
cosine_sim = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
cosine_dist = 1 - cosine_sim

# Visualize
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u = [1, 2]')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='v = [3, 4]')
plt.plot([u[0], v[0]], [u[1], v[1]], 'k--', label='Euclidean path')
plt.plot([u[0], v[0]], [u[1], u[1]], 'g--')
plt.plot([v[0], v[0]], [u[1], v[1]], 'g--', label='Manhattan path')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Distance Metrics Visualization')
plt.legend()
plt.show()

print(f'Euclidean distance: {euclidean_dist:.3f}')
print(f'Manhattan distance: {manhattan_dist:.3f}')
print(f'Cosine distance: {cosine_dist:.3f}')
```

**Visualization Explanation**: The plot shows vectors $\mathbf{u} = [1, 2]$ and $\mathbf{v} = [3, 4]$, with the Euclidean distance as a straight line and Manhattan distance as a grid-like path, highlighting their different approaches to measuring separation.

### 2.3 Norm-Induced Metrics
**Definition**: A norm induces a distance via $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|$, satisfying metric properties.

**Explanation**: Norm-induced distances ensure consistent, mathematically sound measurements, critical for ML algorithms operating in metric spaces.

**Derivation**: The distance $d(\mathbf{u}, \mathbf{v}) = \|\mathbf{u} - \mathbf{v}\|$ inherits norm properties, ensuring non-negativity, symmetry, and the triangle inequality.

**Why Necessary**:
- **Consistency**: Ensures reliable distance measurements.
- **Geometric Algorithms**: Supports clustering, k-NN, and anomaly detection.
- **Optimization**: Provides well-defined objective functions.

**Example**: Euclidean distance is induced by the L2 norm.

**ML Use-Case**: K-means clustering uses Euclidean distance to assign points to centroids.

```python
# Norm-Induced Distance
import numpy as np

# Define vectors
u = np.array([1, 2])
v = np.array([3, 4])

# Compute norm-induced distance
distance = np.linalg.norm(u - v)

print(f'Norm-induced distance (Euclidean): {distance:.3f}')
```

### 2.4 Norms and Orthogonality
**Definition**: Two vectors $\mathbf{u}, \mathbf{v}$ are orthogonal if $\mathbf{u} \cdot \mathbf{v} = 0$. A vector is normalized if $\|\mathbf{u}\|_2 = 1$.

**Explanation**: Orthogonal vectors have zero correlation in direction, and normalization ensures unit length, simplifying projections and similarity measures.

**Derivation**: The dot product $\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\|_2 \|\mathbf{v}\|_2 \cos(\theta)$ is zero when $\theta = 90^\circ$. Normalization divides a vector by its L2 norm.

**Why Necessary**:
- **PCA**: Orthonormal bases simplify projections.
- **Similarity**: Normalized vectors ensure cosine similarity measures angle only.
- **Stability**: Orthogonal directions reduce numerical errors.

**Example**: For $\mathbf{u} = [3, 4]$, $\mathbf{v} = [4, -3]$, their dot product is $3 \cdot 4 + 4 \cdot (-3) = 0$, so they are orthogonal.

**ML Use-Case**: Normalized word embeddings in NLP for cosine similarity.

```python
# Orthogonality Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define orthogonal vectors
u = np.array([3, 4])
v = np.array([4, -3])

# Normalize vectors
u_norm = u / np.linalg.norm(u)
v_norm = v / np.linalg.norm(v)

# Compute dot product
dot_product = np.dot(u_norm, v_norm)

# Visualize
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, u_norm[0], u_norm[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u_norm = [3/5, 4/5]')
plt.quiver(0, 0, v_norm[0], v_norm[1], angles='xy', scale_units='xy', scale=1, color='red', label='v_norm = [4/5, -3/5]')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Orthogonal Normalized Vectors')
plt.legend()
plt.show()

print(f'Dot product (should be ~0): {dot_product:.3e}')
```

### Example 1: k-Nearest Neighbors (Distance-Based Classification)
This example uses Euclidean distance to find the nearest neighbor to a query point.

```python
# k-NN Example
import numpy as np
import matplotlib.pyplot as plt

# Simulated dataset: 3 classes
np.random.seed(0)
class1 = np.random.multivariate_normal([1, 1], [[0.2, 0], [0, 0.2]], 10)
class2 = np.random.multivariate_normal([3, 3], [[0.2, 0], [0, 0.2]], 10)
class3 = np.random.multivariate_normal([2, 4], [[0.2, 0], [0, 0.2]], 10)
data = np.vstack([class1, class2, class3])
labels = np.array([1]*10 + [2]*10 + [3]*10)

# Query point
query = np.array([2, 3])

# Compute distances
distances = np.linalg.norm(data - query, axis=1)
nearest_idx = np.argmin(distances)

# Visualize
plt.figure(figsize=(6, 6))
plt.scatter(class1[:, 0], class1[:, 1], color='blue', label='Class 1')
plt.scatter(class2[:, 0], class2[:, 1], color='green', label='Class 2')
plt.scatter(class3[:, 0], class3[:, 1], color='red', label='Class 3')
plt.scatter(query[0], query[1], color='black', marker='*', s=200, label='Query')
plt.scatter(data[nearest_idx, 0], data[nearest_idx, 1], color='purple', s=200, edgecolor='black', label='Nearest Neighbor')
plt.plot([query[0], data[nearest_idx, 0]], [query[1], data[nearest_idx, 1]], 'k--', label='Distance')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid(True)
plt.title('k-NN: Nearest Neighbor with Euclidean Distance')
plt.legend()
plt.show()

print(f'Nearest neighbor: {data[nearest_idx]}, Class: {labels[nearest_idx]}, Distance: {distances[nearest_idx]:.3f}')
```

**Example Explanation**: This code simulates a k-NN classification task, computing Euclidean distances to find the nearest neighbor to a query point. The visualization shows data points, the query, and the nearest neighbor, with a dashed line indicating the distance.

### Example 2: Linear Regression (Projections and Error Norms)
**Explanation**: In linear regression, the goal is to find the best fit line $y = \beta_0 + \beta_1 x$ that minimizes the sum of squared errors (L2 norm of residuals). This is achieved by projecting the target vector $\mathbf{y}$ onto the column space of the feature matrix $\mathbf{X}$ (including a column of ones for the intercept). The projected points $\mathbf{y}_{\text{proj}} = \mathbf{X} \mathbf{\beta}$ lie on the best fit line, and the residuals $\mathbf{e} = \mathbf{y} - \mathbf{y}_{\text{proj}}$ are orthogonal to the column space of $\mathbf{X}$.

**Formal Derivation**:
- Model: $y_i = \beta_0 + \beta_1 x_i + \epsilon_i$, where $\mathbf{y} = [y_1, \dots, y_n]^T$, $\mathbf{X} = [\mathbf{1}, \mathbf{x}]$, and $\mathbf{\beta} = [\beta_0, \beta_1]^T$.
- Goal: Minimize $\|\mathbf{y} - \mathbf{X} \mathbf{\beta}\|_2^2$.
- The projection of $\mathbf{y}$ onto the column space of $\mathbf{X}$ is $\mathbf{y}_{\text{proj}} = \mathbf{X} \mathbf{\beta}$, where $\mathbf{\beta}$ satisfies the normal equations:
  $$
  \mathbf{X}^T (\mathbf{y} - \mathbf{X} \mathbf{\beta}) = 0 \implies \mathbf{X}^T \mathbf{X} \mathbf{\beta} = \mathbf{X}^T \mathbf{y}
  $$
- Solve for $\mathbf{\beta}$: $\mathbf{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$.
- Projection matrix: $\mathbf{P} = \mathbf{X} (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T$, so $\mathbf{y}_{\text{proj}} = \mathbf{P} \mathbf{y}$.
- Residuals: $\mathbf{e} = \mathbf{y} - \mathbf{y}_{\text{proj}}$, with $\|\mathbf{e}\|_2^2$ as the sum of squared errors (SSE).

**Why Necessary**:
- The projection minimizes the L2 norm of residuals, ensuring the best fit line is closest to the data points in the Euclidean sense.
- Residuals are orthogonal to the column space, ensuring the solution is optimal.

**Solution for Errors**:
- Compute residuals $\mathbf{e}$ and SSE $\|\mathbf{e}\|_2^2$.
- If errors are large, consider:
  - Adding features to $\mathbf{X}$ to expand the column space.
  - Using regularization (e.g., ridge regression) to stabilize $\mathbf{X}^T \mathbf{X}$.
  - Checking for outliers, which may inflate the error norm.

**Example**: Fit a line to points $(x_i, y_i)$ and compute projections and errors.

```python
# Linear Regression: Projection onto Best Fit Line
import numpy as np
import matplotlib.pyplot as plt

# Simulated data
X = np.array([1, 2, 3, 4, 5])  # Feature values
y = np.array([2, 3, 5, 6, 8])  # Target values
X_matrix = np.vstack([np.ones(len(X)), X]).T  # Feature matrix with intercept

# Compute coefficients using normal equations
beta = np.linalg.inv(X_matrix.T @ X_matrix) @ X_matrix.T @ y
beta_0, beta_1 = beta  # Intercept and slope

# Projection matrix
P = X_matrix @ np.linalg.inv(X_matrix.T @ X_matrix) @ X_matrix.T

# Projected y (points on best fit line)
y_proj = P @ y

# Residuals (errors)
errors = y - y_proj
sse = np.sum(errors**2)

# Visualize
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_proj, color='red', label=f'Best Fit Line: y = {beta_0:.2f} + {beta_1:.2f}x')
plt.scatter(X, y_proj, color='green', marker='x', s=100, label='Projected Points')
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_proj[i]], 'k--', label='Residual' if i == 0 else None)
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.title('Linear Regression: Projection onto Best Fit Line')
plt.legend()
plt.show()

print(f'Coefficients: Intercept = {beta_0:.2f}, Slope = {beta_1:.2f}')
print(f'Projected y (points on line): {y_proj.round(2)}')
print(f'Residuals: {errors.round(2)}')
print(f'Sum of squared errors (SSE): {sse:.2f}')
print(f'L2 norm of residuals: {np.linalg.norm(errors):.2f}')
```

**Example Explanation**: This code fits a linear regression model to data points $(x_i, y_i)$ by projecting $\mathbf{y}$ onto the column space of $\mathbf{X}$. The visualization shows:
- Blue dots: Original data points.
- Red line: Best fit line ($y = \beta_0 + \beta_1 x$).
- Green crosses: Projected points ($\mathbf{y}_{\text{proj}}$) on the line.
- Black dashed lines: Residuals ($\mathbf{y} - \mathbf{y}_{\text{proj}}$), orthogonal to the column space.
The output includes coefficients, projected points, residuals, and SSE. To reduce errors, one could add polynomial features, regularize the model, or remove outliers if SSE is high.
