# Equations of Lines, Planes, and Hyperplanes in Linear Algebra for AI/ML

## Introduction and Relevance
**Definition**: Lines, planes, and hyperplanes are geometric objects defined by linear equations in vector spaces, representing one-dimensional, two-dimensional, and (n-1)-dimensional subspaces or affine subspaces, respectively. They are critical in AI/ML for modeling decision boundaries, data relationships, and transformations.

**Mathematical Context**:
- **Line**: A one-dimensional subspace or affine subspace in $\mathbb{R}^n$, defined by a point and a direction vector.
- **Plane**: A two-dimensional subspace or affine subspace in $\mathbb{R}^n$, defined by a point and two direction vectors or a normal vector.
- **Hyperplane**: An (n-1)-dimensional subspace or affine subspace in $\mathbb{R}^n$, defined by $\mathbf{w} \cdot \mathbf{x} + b = 0$, generalizing lines (2D) and planes (3D).

**Mathematical Definitions**:
- **Line**: Parametric form: $\mathbf{x} = \mathbf{p} + t \mathbf{d}$, where $\mathbf{p}$ is a point, $\mathbf{d}$ is the direction vector, and $t \in \mathbb{R}$.
- **Plane**: Parametric form: $\mathbf{x} = \mathbf{p} + s \mathbf{u} + t \mathbf{v}$, where $\mathbf{p}$ is a point, $\mathbf{u}, \mathbf{v}$ are linearly independent direction vectors, and $s, t \in \mathbb{R}$. Normal form: $\mathbf{n} \cdot (\mathbf{x} - \mathbf{p}) = 0$.
- **Hyperplane**: General form: $\mathbf{w} \cdot \mathbf{x} + b = 0$, where $\mathbf{w}$ is the normal vector and $b$ is a scalar. Parametric form: $\mathbf{x} = \mathbf{p} + t_1 \mathbf{v}_1 + \dots + t_{n-1} \mathbf{v}_{n-1}$.

**Examples**:
- In $\mathbb{R}^2$, a line: $y = 2x + 1$ or parametrically $\mathbf{x} = [0, 1] + t [1, 2]$.
- In $\mathbb{R}^3$, a plane: $x + y + z = 1$ or $\mathbf{x} = [1, 0, 0] + s [1, 0, 0] + t [0, 1, 0]$.
- In ML, hyperplanes in SVM: $\mathbf{w} \cdot \mathbf{x} + b = 0$ defines a decision boundary.

**Relevance in Mathematics**:
- **Geometric Modeling**: Represent linear or affine subspaces in $\mathbb{R}^n$.
- **System Solving**: Define solutions to linear equations (e.g., intersection of hyperplanes).
- **Transformations**: Serve as constraints or boundaries in linear transformations.
- **Distance and Projection**: Enable computing distances to subspaces or projecting vectors.

**Relevance in AI/ML**:
- **Decision Boundaries**: Separate classes in classification tasks (e.g., SVM, logistic regression).
- **Model Relationships**: Represent linear relationships in regression or feature spaces.
- **Optimization**: Define constraints in linear programming or ML optimization.
- **Geometric Insights**: Provide interpretations of data structure in high-dimensional spaces.

**Applications Overview**:
- **SVM**: Hyperplanes separate classes in high-dimensional spaces.
- **Linear Regression**: Lines or hyperplanes model feature-target relationships.
- **Logistic Regression**: Hyperplanes define decision boundaries.
- **Computer Vision**: Planes model surfaces in 3D reconstruction.
- **Neural Networks**: Hyperplanes in linear layers transform inputs.
- **Optimization**: Hyperplanes define constraints in ML algorithms.

## Core Concepts and Derivations
This section covers the equations of lines, planes, and hyperplanes, their properties, and their geometric interpretations, with examples, derivations, and visualizations.

### 2.1 Equation of a Line
**Definition**:
- In $\mathbb{R}^2$:
  - Slope-intercept form: $y = mx + c$.
  - General form: $ax + by + c = 0$.
  - Parametric form: $\mathbf{x} = \mathbf{p} + t \mathbf{d}$, $t \in \mathbb{R}$.
- In $\mathbb{R}^n$: Parametric form: $\mathbf{x} = \mathbf{p} + t \mathbf{d}$.

**Explanation**: A line is a one-dimensional affine subspace defined by a point $\mathbf{p}$ and direction vector $\mathbf{d}$. In ML, lines model simple relationships or decision boundaries.

**Derivation**: The parametric form arises from linear combinations: any point on the line is a linear combination of a base point and a direction vector. The general form $ax + by + c = 0$ in $\mathbb{R}^2$ is derived by setting $\mathbf{w} \cdot \mathbf{x} + c = 0$, where $\mathbf{w} = [a, b]$ is normal to the line.

**Why Necessary**:
- **Linear Relationships**: Models relationships in regression or optimization trajectories.
- **Decision Boundaries**: Defines simple boundaries in low-dimensional spaces.
- **Geometric Simplicity**: Provides a basic framework for higher-dimensional generalizations.
- **Data Analysis**: Represents paths or trends in 2D data visualizations.

**Example**: Line in $\mathbb{R}^2$: $\mathbf{x} = [1, 2] + t [2, 3]$.

**ML Use-Case**: In linear regression, a line $y = w_1 x + w_0$ models the relationship between a feature and target.

```python
# Line Visualization
import numpy as np
import matplotlib.pyplot as plt

# Define line: x = p + t * d
p = np.array([1, 2])  # Point
d = np.array([2, 3])  # Direction vector
t = np.linspace(-1, 1, 50)
line_points = p + np.outer(t, d)

# Visualize
plt.figure(figsize=(6, 6))
plt.plot(line_points[:, 0], line_points[:, 1], 'b-', label='Line: x = [1, 2] + t[2, 3]')
plt.scatter(p[0], p[1], color='red', label='Point p = [1, 2]')
plt.quiver(p[0], p[1], d[0], d[1], color='green', scale=10, label='Direction d = [2, 3]')
plt.xlim(-2, 4)
plt.ylim(-1, 6)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Line in $\mathbb{R}^2$')
plt.legend()
plt.show()

print("Points on line:\n", line_points[:5])  # Show first few points
```

### 2.2 Equation of a Plane
**Definition**:
- In $\mathbb{R}^3$:
  - General form: $ax + by + cz + d = 0$.
  - Parametric form: $\mathbf{x} = \mathbf{p} + s \mathbf{u} + t \mathbf{v}$, $s, t \in \mathbb{R}$.
  - Normal form: $\mathbf{n} \cdot (\mathbf{x} - \mathbf{p}) = 0$.

**Explanation**: A plane is a two-dimensional affine subspace defined by a point $\mathbf{p}$ and two linearly independent direction vectors $\mathbf{u}, \mathbf{v}$, or by a normal vector $\mathbf{n}$. In ML, planes model surfaces or decision boundaries in 3D.

**Derivation**: The normal form arises from the orthogonality of the normal vector to the plane’s direction vectors. The general form $\mathbf{w} \cdot \mathbf{x} + d = 0$ is equivalent, where $\mathbf{w} = [a, b, c]$ is the normal vector.

**Why Necessary**:
- **Higher-Dimensional Modeling**: Represents relationships or boundaries in 3D data.
- **Decision Surfaces**: Defines boundaries in 3D classification problems.
- **Geometric Analysis**: Models surfaces in computer vision or 3D data analysis.
- **Generalization**: Extends to hyperplanes in higher dimensions.

**Example**: Plane in $\mathbb{R}^3$: $x + y + z = 1$ or $\mathbf{x} = [1, 0, 0] + s [1, 0, 0] + t [0, 1, 0]$.

**ML Use-Case**: In computer vision, planes model surfaces in 3D point cloud analysis.

```python
# Plane Visualization
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define plane: x = p + s * u + t * v
p = np.array([1, 1, 1])  # Point
u = np.array([1, 0, 0])  # Direction vector 1
v = np.array([0, 1, 0])  # Direction vector 2
s, t = np.meshgrid(np.linspace(-1, 1, 10), np.linspace(-1, 1, 10))
plane_points = p + s[:, :, np.newaxis] * u + t[:, :, np.newaxis] * v

# Compute normal vector
n = np.cross(u, v)  # Normal to plane

# Visualize
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(plane_points[:, :, 0], plane_points[:, :, 1], plane_points[:, :, 2], alpha=0.5, color='blue', label='Plane')
ax.scatter(p[0], p[1], p[2], color='red', s=100, label='Point p = [1, 1, 1]')
ax.quiver(p[0], p[1], p[2], u[0], u[1], u[2], color='green', label='Direction u = [1, 0, 0]')
ax.quiver(p[0], p[1], p[2], v[0], v[1], v[2], color='purple', label='Direction v = [0, 1, 0]')
ax.quiver(p[0], p[1], p[2], n[0], n[1], n[2], color='orange', label='Normal n')
ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
ax.set_zlim(0, 2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane in $\mathbb{R}^3$')
plt.legend()
plt.show()

print("Points on plane (first few):\n", plane_points.reshape(-1, 3)[:5])
print("Normal vector:\n", n)
```

### 2.3 Equation of a Hyperplane
**Definition**:
- In $\mathbb{R}^n$:
  - General form: $\mathbf{w} \cdot \mathbf{x} + b = 0$, where $\mathbf{w}$ is the normal vector and $b$ is a scalar.
  - Parametric form: $\mathbf{x} = \mathbf{p} + t_1 \mathbf{v}_1 + \dots + t_{n-1} \mathbf{v}_{n-1}$, where $\mathbf{v}_i$ are linearly independent.

**Explanation**: A hyperplane is an (n-1)-dimensional affine subspace, generalizing lines and planes. In ML, hyperplanes define decision boundaries in high-dimensional spaces.

**Derivation**: The general form arises from the orthogonality of $\mathbf{w}$ to the hyperplane. The parametric form represents points as linear combinations of a point and direction vectors spanning the hyperplane.

**Why Necessary**:
- **Decision Boundaries**: Defines separating surfaces in classification (e.g., SVM).
- **High-Dimensional Modeling**: Represents relationships in high-dimensional data.
- **Optimization Constraints**: Defines linear constraints in ML optimization.
- **Geometric Insights**: Provides a framework for distances and projections.

**Example**: In $\mathbb{R}^3$, hyperplane: $x - y + z = 2$.

**ML Use-Case**: In SVM, hyperplanes separate classes in high-dimensional feature spaces.

```python
# Hyperplane Visualization (2D slice in 3D)
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define hyperplane: w * x + b = 0
w = np.array([1, -1, 1])  # Normal vector
b = 2  # Intercept

# Create points for visualization (solve for z)
x, y = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10))
z = (-w[0] * x - w[1] * y - b) / w[2]  # From w * x + b = 0

# Visualize
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.5, color='blue', label='Hyperplane')
ax.quiver(0, 0, 0, w[0], w[1], w[2], color='red', label='Normal w = [1, -1, 1]')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-2, 2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Hyperplane in $\mathbb{R}^3$: $x - y + z = 2$')
plt.legend()
plt.show()

# Check if point lies on hyperplane
x = np.array([1, 1, 2])
on_hyperplane = np.abs(np.dot(w, x) + b) < 1e-10
print("Point [1, 1, 2] on hyperplane?:", on_hyperplane)
print("Distance to hyperplane:\n", np.abs(np.dot(w, x) + b) / np.linalg.norm(w))
```

### 2.4 Properties of Lines, Planes, and Hyperplanes
**Definition**:
- **Dimension**: Lines (1D), planes (2D), hyperplanes (n-1)-dimensional in $\mathbb{R}^n$.
- **Normal Vector**: $\mathbf{w}$ in $\mathbf{w} \cdot \mathbf{x} + b = 0$ is perpendicular to the hyperplane.
- **Affine Subspaces**: If $b \neq 0$, the subspace is affine; if $b = 0$, it’s a subspace through the origin.
- **Intersection**: Multiple hyperplanes intersect to form lower-dimensional subspaces (e.g., two planes in $\mathbb{R}^3$ form a line).

**Explanation**: These properties define the geometric structure of lines, planes, and hyperplanes, making them suitable for modeling data relationships or boundaries.

**Why Necessary**:
- **Geometric Framework**: Provides a structure for defining relationships or separations.
- **Distance Computations**: Enables calculating distances to subspaces (e.g., SVM margins).
- **Intersections**: Models solutions to systems of linear equations.
- **ML Applications**: Supports classification, regression, and optimization.

**Example**: Distance to hyperplane $\mathbf{w} \cdot \mathbf{x} + b = 0$ from point $\mathbf{x}_0$: $\frac{|\mathbf{w} \cdot \mathbf{x}_0 + b|}{\|\mathbf{w}\|}$.

**ML Use-Case**: In SVM, the normal vector $\mathbf{w}$ and intercept $b$ define the margin size.

```python
# Distance to Hyperplane
import numpy as np
import matplotlib.pyplot as plt

# Define hyperplane: w * x + b = 0
w = np.array([1, -1])  # Normal vector
b = 2  # Intercept
point = np.array([1, 1])  # Test point

# Compute distance
distance = np.abs(np.dot(w, point) + b) / np.linalg.norm(w)

# Visualize
x = np.linspace(-2, 3, 100)
y = (-w[0] * x - b) / w[1]  # Line equation
plt.figure(figsize=(6, 6))
plt.plot(x, y, 'b-', label='Hyperplane: x - y + 2 = 0')
plt.scatter(point[0], point[1], color='red', label=f'Point [{point[0]}, {point[1]}]')
plt.quiver(point[0], point[1], w[0], w[1], color='green', scale=10, label='Normal w = [1, -1]')
plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title(f'Distance to Hyperplane: {distance:.2f}')
plt.legend()
plt.show()

print("Distance to hyperplane:", distance)
```

## Practical ML Examples
Below are coded examples demonstrating lines, planes, and hyperplanes in ML contexts.

```python
# Example 1: Linear Regression (Line in 2D)
import numpy as np
import matplotlib.pyplot as plt

# Simulate data
X = np.array([[1, 1], [1, 2], [1, 3]])  # Feature matrix with bias term
y = np.array([2, 3, 5])  # Target values

# Normal equation
w = np.linalg.inv(X.T @ X) @ X.T @ y

# Visualize
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 1], y, color='blue', label='Data')
x_vals = np.array([1, 3])
y_vals = w[0] + w[1] * x_vals
plt.plot(x_vals, y_vals, 'r-', label=f'Line: y = {w[1]:.2f}x + {w[0]:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('Linear Regression: Line Fit')
plt.legend()
plt.show()

print("Line equation: y = {:.2f}x + {:.2f}".format(w[1], w[0]))
```

```python
# Example 2: SVM-like Hyperplane (Classification)
import numpy as np
import matplotlib.pyplot as plt

# Simulate data
X = np.array([[1, 1], [2, 2], [3, 1], [1, 3]])  # Features
y = np.array([1, 1, -1, -1])  # Labels (+1 or -1)

# Define hyperplane (example parameters)
w = np.array([1, -1])  # Normal vector
b = -0.5  # Intercept

# Predictions
predictions = np.sign(np.dot(X, w) + b)

# Visualize
x = np.linspace(0, 4, 100)
y_line = (-w[0] * x - b) / w[1]
plt.figure(figsize=(6, 6))
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class +1')
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], color='red', label='Class -1')
plt.plot(x, y_line, 'g-', label='Hyperplane: x - y - 0.5 = 0')
plt.quiver(2, (-w[0] * 2 - b) / w[1], w[0], w[1], color='purple', scale=10, label='Normal w = [1, -1]')
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('SVM-like Hyperplane for Classification')
plt.legend()
plt.show()

print("Predictions:", predictions)
```
